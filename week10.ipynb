```python
num = int(input("1) 암호화  2) 복호화\n메뉴를 선택하세요(1,2): "))

if(num == 1):
    a = input("input: ")
    b = a.split(" ")
    c = b[0]
    d = b[1]
    x = ""
    y = ""
    
    for i in range(0, len(d), 1):
        x += chr(ord(d[i]) + 1)
        
    for i in range(0, len(c), 1):
        y += chr(ord(c[i]) + 1)
        
    print("print: " + x + " " + y)
    
if(num == 2):
    a = input("input: ")
    b = a.split(" ")
    c = b[0]
    d = b[1]
    x = ""
    y = ""
    
    for i in range(0, len(d), 1):
        x += chr(ord(d[i]) - 1)
        
    for i in range(0, len(c), 1):
        y += chr(ord(c[i]) - 1)
        
    print("print: " + x + " " + y)
```

```python
td = {}

for i in range(10):
    n = input("이름: ")
    p = input("전화번호: ")
    td[n] = p

while 1 :
    menuSelect = input("\n1) 이름으로 검색  2) 전화번호로 검색  3) 종료\n메뉴를 선택하세요(1~3): ")
    
    if menuSelect == "1":
        sn = input("이름: ")
        
        for n, p in td.items():
            if n == sn:
                print("전화번호: " + p)
    elif menuSelect == "2":
        sp = input("전화번호: ")
        
        for n, p in td.items():
            if p == sp:
                print("이름: " + n)
    elif menuSelect == "3":
        print("프로그램이 종료되었습니다.")
        break;
    else:
        print("올바른 값을 입력해주세요.")
```
